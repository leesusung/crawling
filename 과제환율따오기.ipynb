```python
data6=response3.json()
```

```python
naver_1=pd.DataFrame(data6)[['name','closePrice']]
```

```python
naver_1['closePrice']=naver_1['closePrice'].apply(lambda naver_1 : float(naver_1.replace(",", "")))
```

```python
# apply 연습
naver_1['hi1']=naver_1['closePrice'].apply(lambda naver_1 : round(naver_1))
naver_1['hi2']=naver_1['closePrice'].apply(lambda naver_1 : 100)
```

```python


#크롤링 함ㅅ구만ㄴ들기

def hello(page, pagesize) :
    url4= f'https://api.stock.naver.com/marketindex/exchange/FX_USDKRW/prices?page={page}&pageSize={pagesize}'
    header1 = {'referer': 'https://m.stock.naver.com/marketindex/exchange/FX_USDKRW', 
                 'user-agent' :'Mozilla/5.0 (iPhone; CPU iPhone OS 13_2_3 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.0.3 Mobile/15E148 Safari/604.1'
    }
    response4=requests.get(url4,headers= header1)
    data4=response4.json()
    return pd.DataFrame(data4)[['localTradedAt','closePrice']]
```

```python
hello(1,40)
```
