```python
# 3 . HTmL (str) >BeautifulSoup Objecct  
```

```python
dom = BeautifulSoup(response.text, 'html.parser')
type(dom)
```

```python
# 4 . BeautifulSoup Objecct  > css  selectir  > Data Frmame 
```

```python
#  200ro 상품 엘리먼트선택
elements = dom.select('#gBestWrap > div > div:nth-child(5) > div > ul > li')
len(elements)
```

```python
#gBestWrap > div > div:nth-child(5) > div > ul > li:nth-child(2)
#gBestWrap > div > div:nth-child(5) > div > ul > li:nth-child(1)
#gBestWrap > div > div:nth-child(5) > div > ul > li:nth-child(3)
element = elements[0]
element
```

```python
data = {
    'title' : element.select_one('.itemname').text,
    'o_price' : element.select_one('.o-price').text,
    's_price' : element.select_one('.s-price> strong').text,
    'image': "http:"+element.select_one('img').get('data-original'),
 }
data
```

```python
items = []
for element in elements:
    data = {
    'title' : element.select_one('.itemname').text,
    'o_price' : element.select_one('.o-price').text,
    's_price' : element.select_one('.s-price> strong').text,
    'image': "http:"+element.select_one('img').get('data-original'),
    }
    items.append(data)
```
